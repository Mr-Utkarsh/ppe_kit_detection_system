```python
"""
Evaluation script for the Object Detection System.
This script calculates the mean Average Precision (mAP) for our custom trained model.
"""

import os
import sys
from ultralytics import YOLO

# Ensure correct working directory so we can find the model directory
current_dir = os.getcwd()
if current_dir.endswith('src'):
    os.chdir('..')

# Define the path to our best generated weights
custom_weights_path = "model/construction_train/weights/best.pt"

# Check if the model has actually been trained before trying to evaluate it
if not os.path.exists(custom_weights_path):
    print(f"Error: The custom weights could not be found at '{custom_weights_path}'.")
    print("Please ensure you run train.ipynb to completion before evaluating.")
    sys.exit(1)

print(f"Successfully located weights at {custom_weights_path}")
print("Loading model into memory...")
model = YOLO(custom_weights_path)

print("Running validation loop on the test dataset...")
# Calculate metrics
metrics = model.val()

# Output the final industry-standard metrics
print("\n==============================")
print("     Validation Metrics       ")
print("==============================")
print(f"mAP50 Accuracy: {metrics.box.map50 * 100:.2f}%")
print(f"mAP50-95 Accuracy: {metrics.box.map * 100:.2f}%")
```
