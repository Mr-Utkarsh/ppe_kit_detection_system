```python
import os
from ultralytics import YOLO

# Since this notebook is in the 'src' folder, we use '../' to go back to the main folder
weights_path = '../model/best.pt'
data_path = '../construction_dataset/data.yaml'

# Check if the model has been trained yet
if os.path.exists(weights_path) == False:
    print('Error: Model weights not found!')
    print('Please run train.ipynb first so the model can be trained and saved.')
else:
    print('Loading the trained model...')
    model = YOLO(weights_path)
    
    print('Evaluating the model...')
    metrics = model.val(data=data_path)
    
    # Extract the accuracy metrics and format them as percentages
    precision = metrics.box.mp * 100
    recall = metrics.box.mr * 100
    map50 = metrics.box.map50 * 100
    map50_95 = metrics.box.map * 100
    
    print(f'Model Precision (Accuracy when it guesses): {precision:.2f}%')
    print(f'Model Recall (How many objects it found): {recall:.2f}%')
    print(f'mAP50 Accuracy (Overall bounding box score): {map50:.2f}%')
    print(f'mAP50-95 Accuracy (Strict bounding box score): {map50_95:.2f}%')
```
