```python
"""
Training script for the Object Detection System.
This script fine-tunes the pre-trained YOLOv8 nano model on our custom construction dataset.
"""

import os
from ultralytics import YOLO

# Adjust the working directory if the script is run directly from the src/ folder
current_dir = os.getcwd()
if current_dir.endswith('src'):
    os.chdir('..')

print("Step 1: Loading the base YOLOv8 model for transfer learning...")
# Initialize the base model
model = YOLO("model/yolov8n.pt")

print("Step 2: Starting the training process on the construction dataset...")
# Train the model with hyperparameters optimized for our hardware
model.train(
    data="construction_dataset/data.yaml", # Path to training data config
    epochs=50,                             # Number of training loops
    imgsz=640,                             # Image size for resizing
    project="model",                       # Directory to save the outputs
    name="construction_train"              # Name of the current training run
)

print("Training successfully completed! Model weights are saved in the model/ directory.")
```
